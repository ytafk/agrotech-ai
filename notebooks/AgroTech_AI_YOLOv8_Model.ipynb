Modelin farklı veri setindeki tespit ve güven skorları aşağıdaki hücrede gözükmektedir. Model eğitim setinde bulunan elma, muz ve portakal sınflarını yüzde 90 üzeri güven skoru ile doğru tespit etmiştir.

```python
results = model.predict(source="/content/Rotten-Fruit-Detector-1/valid/images", conf=0.25)

for r in results:

    filename = r.path.split('/')[-1]

    for box in r.boxes:
        class_id = int(box.cls[0])
        class_name = model.names[class_id]
        conf_score = float(box.conf[0])

        print(f"Görsel: {filename} | Tespit: {class_name} | Güven: %{conf_score*100:.2f}")
```
